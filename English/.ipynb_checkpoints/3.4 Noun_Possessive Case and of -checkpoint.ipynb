- Если человек или животное – мы используем правила -s 
- Если здания или предметы – мы используем of-phrase






**Притяжательные неодушевленные предметы**

• Неодушевленные слова, представляющие собой совокупность одушевленных существ  
например, прибыль компании (традиционная)

• Час, минута и тд.  
например, двухчасовая работа

• 's возможно  
например, на современных компьютерах  
подсчет ссылок на объект



***



**Кластер существительных (STE)**

• В английском языке одно или несколько существительных  
могут быть использованы для модификации сущ. (кластер существительных)

• Не делайте кластеры существительных более чем из 3 существительных

например,  
=>  _Соединение света взлетно-посадочной полосы (короткое)_  
=>  ~~_Калибровка сопротивления соединения взлетно-посадочных полос_~~ _(слишком длинное)_



***



**Руководство по стилю MS рекомендует:**

• Не используйте притяжательную форму:  
другие названия компаний, если у вас нет другого выбора;  
название продукта, услуги или функции

• Вы можете использовать эти имена в качестве прилагательных  
или вместо них использовать **конструкцию "of"**:  

Например,  
~~интерфейс Windows  
Шаблоны Word  
словарь проверки орфографии~~
