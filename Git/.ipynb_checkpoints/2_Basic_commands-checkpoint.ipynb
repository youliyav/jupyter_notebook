### Файлы в git, можно разделить на три группы:

- Непроиндексированные файлы в локальном хранилище
- Проиндексированные файлы в локальном репозитории
- Проиндексированные файлы на удалённом сервере


> **Проиндексированный файл - это файл, за которым система git умеет следить:  
понимает, когда у него изменилось содержимое, имя, когда он удалён и прочее.**

Когда в файловой системе создаётся новый файл, система git о нём ничего не знает до тех пор,  
пока он не будет проиндексирован.  

После индексации файл можно даже не отправлять на удалённый сервер,  
git уже будет хранить историю изменений этого файла.  

Сохранение файла на удалённом сервере полезна, когда над проектом работают несколько человек,  
и есть необходимость обмениваться файлами.

> **"git add"** индексирует файлы  
**"git commit"** фиксирует изменения  
**"git push"** отправляет на удалённый сервер

**Переводить файлы можно из состояния 1 в 2, 2 в 3 и 3 в 2. Остальные операции невозможны**  
   _Невозможно добавить файл в индекс только на удалённом сервере,  
   не проиндексировав его в локальном репозитории (прямой переход 1 в 3 невозможен).  
   Также нельзя из состояний 2 или 3 перевести в состояние 1._
 

> **_GIT - СИСТЕМА КОНТРОЛЯ (управляется командами из консоли, хранит историю изменений файлов)_**  


> **_GITHUB - файловое хранилище_**




#### _Копируем SSH url_



#### _Используем клонирование (git clone и SSH url)_





#### _Переходим в репозиторий git-example (cd) и проверяем содержимое папки (ll)_



#### _Открываем файл .md через vim и редактируем его_



#### _Создаем новый файл .txt (touch)_



#### _После команды git add <название файла> - файл индексируется. Содержимое рабочей директории добавляется  в индекс (staging area) для последующего коммита)_



#### _Делаем git commit -am "Комментарий" - команда для записи индексированных изменений в репозиторий_



#### _Делаем git push origin master для выгрузки содержимого локального репозитория в удаленный репозиторий_


```
git clone SSH url
cd git-example
ll
git vim README.md 
git touch newfile.txt
git add newfile.txt
git commit -am "New version of repo"
git push origin master
```
