## Получение ключа SSH-key


#### _Если нет зарегистрированного ключа SSH, то получаем ошибку_




***
##  _Создаем ключ "ssh-keygen -t rsa" 
_Пароли не вводим, просто нажимаем Enter_






***
## _Далее команда "cat ~/.ssh/id_rsa.pub" генерирует ключ_






***
## _Регистрируем ключ [здесь](https://github.com/settings/ssh/new)_






## _Нажимаем "add ssh key" и вводим пароль от аккаунта github_
