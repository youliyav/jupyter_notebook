# Задача

В директории  /home/box/REPO у вас содержится склонированный репозиторий https://github.com/OSLL/git_course_example_repo.

В репозитории находится файл file.txt, который менялся на протяжении 4 коммитов. Вам необходимо найти, какие данные были добавлены в file.txt в третьем коммите и удалить остальные (т.е. нужно оставить только строку, которая была добавлена во время третьего коммита).


Ответ: 

```
1 вариант:

git log file.txt
git show c709ddd30edeb846a1ce3a811de9b8a8553f6f2a
echo oPL3g1jH9t6iUvNJi8QnWzWzajBqUgnx > file.txt


2 вариант:

git log --oneline file.txt  # выводится сокращенные хэши и сообщение коммитов
git diff d2b3ce3 a8e41c2  # < 2 commit > < 3 commit > сравниваем коммиты и показываем изменения 
echo 'oPL3g1jH9t6iUvNJi8QnWzWzajBqUgnx' > file.txt  # перезаписываем файл (строку сохраняем в file.txt)


```
