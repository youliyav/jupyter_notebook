**Слово "инкапсуляция" происходит от латинского in capsula** —  
"размещение в оболочке". Таким образом, инкапсуляцию можно интуитивно понимать как изоляцию

**Инкапсуляция — это:**
- ограничение доступа к составляющим объект компонентам (методам и переменным).  
- делает некоторые из компонент доступными только внутри класса.  
- в Python работает лишь на уровне соглашения между программистами о том,  
какие атрибуты являются общедоступными, а какие — внутренними.




```Python
class Person:
    def __init__(self, name, surname):
        self._name = name                            # приватный 
        self._surname = surname                      # приватный 
        self.name = f'{self._name} {self._surname}'  # публичный


p = Person('Julia', 'Bratishko')
print(p.name)
```

`# приватный метод (исп. внутри класса)`
